```julia
function MA(X, n)
    S = Float64[]
    for i in 1 : length(X) - n + 1
        s = mean(X[i : i + n - 1])
        append!(S, s)
    end
    return S
end
```

```julia
scatter(MA(cars, 2))
```

```julia
scatter(MA(cars, 3))
```

```julia
scatter(MA(cars, 4))
```
